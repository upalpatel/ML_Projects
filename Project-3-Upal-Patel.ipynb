Question 1:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, metrics
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score 

df1 = pd.read_csv('dataset1.csv')

x1 = df1.iloc[:, [0,1]].values
y1 = df1.iloc[:, 2].values

#split the data into training and testing sets to test accuracy
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size = 0.2, random_state = 0)
```

```python
#create svm for linear kernel
linear_svc = svm.SVC(kernel = 'linear', gamma = 'scale')

linear_svc.fit(x1_train, y1_train)
y1_pred = linear_svc.predict(x1_test)
print("linear kernel score:", accuracy_score(y1_test, y1_pred))
```

Question 2:

```python
X1, X2 = np.meshgrid(np.arange(start = x1[:, 0].min() - 1,  
                               stop = x1[:, 0].max() + 1, step = 0.01), 
                     np.arange(start = x1[:, 1].min() - 1,  
                               stop = x1[:, 1].max() + 1, step = 0.01)) 
  
plt.contourf(X1, X2, linear_svc.predict( 
             np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), 
             alpha = 0.8, cmap = ListedColormap(('white', 'lightgray'))) 
  
plt.xlim(X1.min(), X1.max()) 
plt.ylim(X2.min(), X2.max()) 

for i, j in enumerate(np.unique(y1)): 
    plt.scatter(x1[y1 == j, 0], x1[y1 == j, 1], 
                c = np.array(ListedColormap(('maroon', 'blue'))(i)).reshape(1,-1), label = j) 

plt.title('SVC with linear kernel - Dataset 1')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()

#dataset 3 with grid search

poly_svc3 = svm.SVC(kernel = 'poly')

#scale the features
x3_train = SS.fit_transform(x3_train)  
x3_test = SS.transform(x3_test)

poly_svc3.fit(x3_train, y3_train)
y3_pred = poly_svc3.predict(x3_test)

cmat3 = confusion_matrix(y3_test, y3_pred)

param_grid = {
    'C': [0.001, 0.01, 0.1, 1.0], 
    'degree': [3, 5, 7],
    'gamma': [0.01, 0.1, 1.0], 
    'kernel': ['poly']}

gridSearch = GridSearchCV(poly_svc3, param_grid, cv=10)

gridSearch.fit(x3_train, y3_train)
gridPredict = gridSearch.predict(x3_test)

cmatGrid = confusion_matrix(y3_test, gridPredict)

print("poly kernel score:", accuracy_score(y3_test, y3_pred))
print("confusion matrix:\n", cmat3)
print("poly kernel score with grid search:", accuracy_score(y3_test, gridPredict))
print("confusion matrix with grid search:\n", cmatGrid)
```
